Linear equation example:
pattern term is 1 and term value is 3
pattern term is 2 and term value is 5
pattern rule is: start at 3, increase by 2 each time
constant rate is +2 each time
relationship equation: constant rate multiplyed by 'n'
n= term number and v=value
make any adjustments
for example:
v=2n+1


```python
import matplotlib.pyplot as plt
import math
```

```python
x = [1, 2, 3, 4, 5]
y = [6*n - 5 for n in x]
```

```python
y
```

```python
plt.plot(x, y, marker='o')
plt.xlabel('term number')
plt.ylabel('term value')
plt.title('Plot of y = 6x - 5')
plt.grid()
plt.xlim(0,5)
plt.show()
```
