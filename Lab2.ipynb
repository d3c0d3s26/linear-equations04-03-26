laptop battery drains faster than usual and I would like to use a plot to understand how much time I have until battery dies.

my battery life starts the day at `100%`.
each hour, I notice the battery decreases by `15%-20%` when using browers.
assume the average number is `18%` per hour.
logic:
x= term number
y= term value
v= value
100= start value

hour  battery
x(0) y(100)
x(1) y(82)
x(2) y(64)

thinking:
v=start + (-18x)
value = start value plus the constant rate multiplyed by the hour

equation:
y=100+(-18x)

here is the final math equation:
$y=100+(-18x)$


```python
import matplotlib.pyplot as plt
```

```python
x = [0, 1, 2, 3, 4, 5, 6, 7]
y = [100+(-18*n) for n in x]
```

```python
x[3]
y[3]
```

```python
plt.plot(x, y, marker='o')
for i in range(len(x)):
    plt.text(x[i], y[i], f'{y[i]}%', va='bottom')
plt.xlabel('term number (hour)')
plt.ylabel('term value (battery percentage)')
plt.title('Plot of y = 100+(-18x)')
plt.grid()
plt.xlim(0,7)
plt.ylim(0,100)
plt.show()
```
